BINNING

```python
def binning(data):
    maximum_value=max(data)
    minimum_value=min(data)
    print(maximum_value,minimum_value)
    #child,teen,adult,old
    
    child=[]
    teen=[]
    adult=[]
    old=[]
    
    for i in data:
        if i<=12:
            child.append(i)
        elif i>12 and i<20:
            teen.append(i)
        elif i>=20 and i<45:
            adult.append(i)
        elif i>45:
            old.append(i)
    print('bin of children=',child)
    print('bin of teen=',teen)
    print('bin of adult=',adult)
    print('bin of old=',old)
            
            
age=[10,82,1,15,4,6,33]
binning(age)
```

```python
def binning(data):
    max_value=max(data)
    min_value=min(data)
    print('max price=',max_value,'min price=',min_value)
    
    
    #low,middle,high
    low=[]
    middle=[]
    high=[]
    
    for i in data:
        if i<=30:
            low.append(i)
        elif i>30 and i<=60:
            middle.append(i)
        elif i>60:
            high.append(i)
    print('bin of low prices=',low)
    print('bin of middle prices=',middle)
    print('bin of high prices=',high)


prices=[55,5,25,70,85,99,6,45,10] 
binning(prices)
```
